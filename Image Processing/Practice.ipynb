```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# image load
image=cv.imread('rifat.jpg')
img_rgb=cv.cvtColor(image,cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
array=[[1,1,1],
        [1,1,1],
        [1,1,1]]
# create a karnel
karnel=np.array(array,dtype=np.float32)
print(karnel)
```

```python
# Now using filter2D
kernel1=np.ones((3,3),dtype=np.float32)/9
blure_image1=cv.filter2D(img_rgb,-1,kernel1)

# using secoind kernel
kernel2=np.ones((25,25),dtype=np.float32)/(25*25)
blure_image2=cv.filter2D(img_rgb,-1,kernel2)
plt.figure(figsize=(15, 5))
plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.subplot(1,3,2)
plt.imshow(blure_image1)
plt.subplot(1,3,3)
plt.imshow(blure_image2)
```

```python
# using buld function
blure_image3=cv.blur(img_rgb,(3,3))
blure_image4=cv.blur(img_rgb,(7,7))
plt.figure(figsize=(15, 5))
plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.subplot(1,3,2)
plt.imshow(blure_image3)
plt.subplot(1,3,3)
plt.imshow(blure_image4)
```
