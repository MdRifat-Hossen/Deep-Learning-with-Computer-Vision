#### Image Thresholding / Simple Thresholding


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image=cv.imread('images.jpg')
img=cv.cvtColor(image,cv.COLOR_RGB2GRAY)
plt.imshow(img,cmap='gray')
```

```python
res,thres=cv.threshold(img,127,126,cv.THRESH_BINARY)
plt.imshow(thres,cmap='gray')
print(res)
```

```python
res,thres=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
plt.imshow(thres ,cmap='gray')
print(res)
```

```python
res,thres=cv.threshold(img,127,255,cv.THRESH_TRUNC)
plt.imshow(thres ,cmap='gray')
print(res)
```
