## Image Thresholding

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('/content/drive/MyDrive/Colab Notebooks/image_processing_batch-2/images/orange-1.jpg', cv.IMREAD_GRAYSCALE)
```

### Global Thresholding

```python
plt.imshow(img, cmap='gray')
```

```python
ret, thresholded = cv.threshold(img, 230, 255, cv.THRESH_BINARY)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret, thresholded = cv.threshold(img, 230, 0, cv.THRESH_BINARY_INV)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret, thresholded = cv.threshold(img, 220, 255, cv.THRESH_TRUNC)
plt.imshow(thresholded, cmap='gray')
print(ret)
```
