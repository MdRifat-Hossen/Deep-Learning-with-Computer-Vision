```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
# Load the image
img=cv.imread('download.jpg')
img_rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
# using build funciton for filtaring
blure_img=cv.blur(img_rgb,(3,3))
blur_imag=cv.blur(img_rgb,(7,7))

# Ploting this imge

plt.figure(figsize=(10,15))
plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')


plt.subplot(1,3,2)
plt.imshow(blure_img)
plt.title("Blur image 3*3")
plt.axis('off')


plt.subplot(1,3,3)
plt.imshow(blur_imag)
plt.title("blure 7*7 Image")
plt.axis('off')
```

```python
kernel=np.ones((3,3),dtype=np.float32)/9 # using the 3*3 kernel
kernel1=np.ones((7,7),dtype=np.float32)/49 # using the 5*5 kernel

blur1=cv.filter2D(img_rgb,-1,kernel)
blur2=cv.filter2D(img_rgb,-1,kernel1)

plt.Figure(figsize=(10,15))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')


plt.subplot(1,3,2)
plt.imshow(blur1)
plt.title("Blur image 3*3")
plt.axis('off')


plt.subplot(1,3,3)
plt.imshow(blur2)
plt.title("blure 7*7 Image")
plt.axis('off')
```
