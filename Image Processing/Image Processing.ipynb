# Changing Color-Space in OpenCV

## Introduction

Color space conversion is a crucial part of image processing, allowing for the manipulation and analysis of images in different color models. OpenCV provides two important functions for this purpose: `cv.cvtColor()` and `cv.inRange()`.

## `cv.cvtColor()`

The `cv.cvtColor()` function is used to convert an image from one color space to another. It is a versatile function that supports a wide range of color conversions.

### Syntax

```python
cv.cvtColor(src, code)


````python
# colorspacing the image 
# every image have convert befor RGB
gray_image=cv.cvtColor(image,cv.COLOR_BGR2GRAY) #when copy the path then time convert into BGR
hvs_image=cv.cvtColor(image,cv.COLOR_BGR2HSV)
rgb_image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
# now ploting this images
plt.figure(figsize=(15,25))
plt.subplot(1,4,1)
plt.imshow(image)
plt.title("Original")
plt.subplot(1,4,2)
plt.imshow(gray_image,cmap='gray')
plt.title("Gray Image")
plt.subplot(1,4,3)
plt.imshow(rgb_image)
plt.title("RGB Image")
plt.subplot(1,4,4)
plt.imshow(hvs_image)
plt.title("Hvs Image")
````

### Rotating 
Have to mathode to rotate the image
1. using `cv2.rotate()` mathode 
Syntax: `cv2.cv.rotate( src, rotateCode[, dst] )`
1. Parameters: src: It is the image whose color space is to be changed.
2. rotateCode: It is an enum to specify how to rotate the array.
3. dst: It is the output image of the same size and depth as 4. src image. It is an optional parameter.
5. Return Value: It returns an image.

````python
im=cv.rotate(image,cv.ROTATE_90_CLOCKWISE)
im1=cv.rotate(image,cv.ROTATE_180)
plt.imshow(im)
plt.imshow(im1)
````

`cv2.getRotationMatrix2D()` function is used to make the transformation matrix M which will be used for rotating a image.

`cv2.getRotationMatrix2D(center, angle, scale)`

center: Center of rotation

angle(θ): Angle of Rotation. Angle is positive for anti-clockwise and negative for clockwise.

scale: scaling factor which scales the image

## Resizing and Recalling
1. Resizing `cv2.resize(source, dsize, dest, fx, fy, interpolation)`
2. Rescalling `cv2.resize(source, dsize, dest, fx, fy, interpolation)` same to Resizing but must be matain ratio

always dest `non`

some Interpolation-
`cv2.INTER_AREA`: This is used when we need to shrink an image.
`cv2.INTER_CUBIC`: This is slow but more efficient.
`cv2.INTER_LINEAR`: This is primarily used when zooming is required. This is the default interpolation technique in OpenCV.

````python
im2=cv.resize(image,None,fx=1,fy=.5,interpolation=cv.INTER_CUBIC)
im3=cv.resize(image,None,fx=.5,fy=.5,interpolation=cv.INTER_CUBIC)
im4=cv.resize(image,None,fx=.7,fy=.5,interpolation=cv.INTER_CUBIC)
plt.subplot(2,3,1)
plt.imshow(im2)
plt.subplot(2,3,2)
plt.imshow(im3)
plt.subplot(2,3,3)
plt.imshow(im4)
im5=cv.resize(image,None,fx=1,fy=1,interpolation=cv.INTER_CUBIC)
im6=cv.resize(image,None,fx=2,fy=2,interpolation=cv.INTER_CUBIC)
im7=cv.resize(image,None,fx=3,fy=3,interpolation=cv.INTER_CUBIC)
plt.subplot(2,3,4)
plt.imshow(im5)
plt.subplot(2,3,5)
plt.imshow(im6)
plt.subplot(2,3,6)
plt.imshow(im7)
````

## Fliping
Syntax: `cv2.cv.flip(src, flipCode[, dst] )`

Parameters: src: Input array.

dst: Output array of the same size and type as src.

If the flipCode is set to `0`, the image is flipped across the x-axis 

and if the flipCode is set to a positive integer (say `1`), the image is flipped across the Y-axis. 

If the flipCode is set to a negative integer (say "`-1`"), the image is flipped across both axes.

````python
plt.imshow(image)
````

````python
# now fliping the image
im1=cv.flip(image,0)
im2=cv.flip(image,1)
im3=cv.flip(image,-1)
plt.subplot(1,4,1)
plt.imshow(image)
plt.subplot(1,4,2)
plt.imshow(im1)
plt.subplot(1,4,3)
plt.imshow(im2)
plt.subplot(1,4,4)
plt.imshow(im3)
````
