````python
import matplotlib.pyplot as plt
import cv2 as cv
````

# Image Read

````python
image_path='images.jpg'
image=cv.imread(image_path)
plt.figure(figsize=(2,7))
plt.imshow(image)
````

# Changing Color-Space in OpenCV

## Introduction

Color space conversion is a crucial part of image processing, allowing for the manipulation and analysis of images in different color models. OpenCV provides two important functions for this purpose: `cv.cvtColor()` and `cv.inRange()`.

## `cv.cvtColor()`

The `cv.cvtColor()` function is used to convert an image from one color space to another. It is a versatile function that supports a wide range of color conversions.

### Syntax

```python
cv.cvtColor(src, code)


````python
# colorspacing the image 
# every image have convert befor RGB
gray_image=cv.cvtColor(image,cv.COLOR_BGR2GRAY) #when copy the path then time convert into BGR
hvs_image=cv.cvtColor(image,cv.COLOR_BGR2HSV)
rgb_image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
# now ploting this images
plt.figure(figsize=(15,25))
plt.subplot(1,4,1)
plt.imshow(image)
plt.title("Original")
plt.subplot(1,4,2)
plt.imshow(gray_image,cmap='gray')
plt.title("Gray Image")
plt.subplot(1,4,3)
plt.imshow(rgb_image)
plt.title("RGB Image")
plt.subplot(1,4,4)
plt.imshow(hvs_image)
plt.title("Hvs Image")
````
